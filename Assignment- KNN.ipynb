```python
var=['mainroad','guestroom', 'basement', 'hotwaterheating', 'airconditioning','prefarea']
house_data[var]=house_data[var].replace(to_replace={'yes':1,'no':0})
house_data.head()
```

```python
house_data['furnishingstatus']=house_data['furnishingstatus'].replace(to_replace={'furnished':1,'semi-furnished':2,'unfurnished':0})
house_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train,test=train_test_split(house_data,train_size=0.7,test_size=0.3,random_state=100)
train.head()
```

```python
train.shape
```

```python
test.shape
```

```python
x_train = train.drop('price', axis=1)
y_train = train['price']

x_test = test.drop('price', axis=1)
y_test = test['price']
```

# Knn 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
```

```python
rmse=[] 
for i in range(0,8):
    i = i+1
    knn = KNeighborsRegressor(n_neighbors=i)

    knn.fit(x_train,y_train)  
    pred=knn.predict(x_test) 

    rms =np.sqrt(mean_squared_error(y_test,pred)) 
    rmse.append(rms)
    print("root_mean_square_error for k=",i,'is:',rms)
    
     
```

The minimum root mean square value is minimum here for k=4,5,6,7,8
 Hence,we have RMSE is 1607419.70834821 for k=5
